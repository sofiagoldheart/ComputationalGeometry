```python
# librerias
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as mtri
import triangle as tr
```

```python
# Aquí tu código
A = tr.get_data('A')

tr_A = tr.triangulate(A, 'a0.0015')

tr.compare(plt, A, tr_A)

plt.show()
```

```python
# Aquí tu código
ell = tr.get_data('ell')

tr_ell = tr.triangulate(ell, 'pc')

tr.compare(plt, ell, tr_ell)

plt.show()
```

```python
# Aquí tu código
spiral = tr.get_data('spiral')

tr_spiral = tr.triangulate(spiral, 'a0.060')

tr.compare(plt, spiral, tr_spiral)

plt.show()
```

```python
# Aquí tu código

land = tr.get_data('greenland')

tr_land = tr.triangulate(land, 'a')

tr.compare(plt, land, tr_land)

plt.show()
```
