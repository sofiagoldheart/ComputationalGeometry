##Introducción

En este trabajo en equipo, presentamos la implementación del algoritmo incremental para encontrar la poligonal de tres polígonos simples creados con la librería de polygenerator. En este reporte, presentamos el código implementado en Python.

```python
from polygenerator import ( #Librerias para crear el polígono
random_polygon,
random_star_shaped_polygon,
random_convex_polygon,
)
import matplotlib.pyplot as plt
import random
import matplotlib.image as mpimg #Para crear imagenes
from scipy.spatial import ConvexHull #Para detectar el poligono convexo
```

```python
#Función para crear el polígono
def crear_poli(polig, archivo):
  plt.figure()
  plt.gca().set_aspect("equal")
  for i, (x, y) in enumerate(polig):
    plt.text(x, y, str(i), horizontalalignment="center", verticalalignment="center")#Generamos los puntos
  polig.append(polig[0])
  xs, ys = zip(*polig)
  plt.plot(xs, ys, "r-", linewidth=0.3)
  plt.savefig(archivo, dpi=300)#Guardamos la figura
  plt.close()
#Función para el algoritmo incremental para encontrar la poligonal
def poli_convex(points):
  hull = ConvexHull(points)
  vertices = [points[i] for i in hull.vertices]
  return vertices
```

###Polígonos


```python
#Primer Polígono
polig = random_polygon(num_points=15)
crear_poli(polig, "random_polygon.png") # Se llama a la función creada
#Mostrar la imagen
figura = mpimg.imread('/content/random_polygon.png')
#Graficar poligono
print("Poligono generado aleatoriamente")
plt.imshow(figura)
plt.show()
```

```python
#Segundo Polígono
polig = random_star_shaped_polygon(num_points=20)
crear_poli(polig, "random_star_shaped_polygon.png")
#Mostrar la imagen
figura = mpimg.imread('/content/random_star_shaped_polygon.png')
#Graficar poligono
print("Poligono generado aleatoriamente")
plt.imshow(figura)
plt.show()
```

```python
#Tercer Polígono
polig = random_convex_polygon(num_points=25)
crear_poli(polig, "random_convex_polygon.png")
#Mostrar la imagen
figura = mpimg.imread('/content/random_convex_polygon.png')
#Graficar poligono
print("Poligono generado aleatoriamente")
plt.imshow(figura)
plt.show()
```

###Encontrar la poligonial

```python
#Primer Polígono
p1 = poli_convex(polig)
crear_poli(p1, "random_polygon_convex.png") # Se llama a la función creada
#Mostrar la imagen
figura = mpimg.imread('/content/random_polygon_convex.png')
#Graficar poligono convexo
print("Poligono Convexo")
plt.imshow(figura)
plt.show()
```

```python
#Segundo Polígono
p2 = poli_convex(polig)
crear_poli(p2, "random_star_shaped_polygon_convex.png") # Se llama a la función creada
# leer la imagen
figura = mpimg.imread('/content/random_star_shaped_polygon_convex.png')
# Graficar poligono convexo
print("Poligono Convexo")
plt.imshow(figura)
plt.show()
```

```python
#Tercera Polígono
p3 = poli_convex(polig)
crear_poli(p3, "random_convex_polygon_convex.png") # Se llama a la función creada
# leer la imagen
figura = mpimg.imread('/content/random_convex_polygon_convex.png')
# Graficar poligono convexo
print("Poligono Convexo")
plt.imshow(figura)
plt.show()
```
