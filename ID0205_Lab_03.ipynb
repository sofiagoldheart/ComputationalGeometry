```python
#Librerias
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

##Puntos en el plano cartesiano
#Definimos los puntos.

## Plano Cartesiano
ancho = 10
alto = 10
fig, ax = plt.subplots(figsize=(ancho, alto))

xmin, xmax, ymin, ymax = -10, 10, -10, 10
ax.set(xlim=(xmin, xmax), ylim=(ymin, ymax), aspect='equal')

ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_xlabel('x', size=14, labelpad=-24, x=1.03)
ax.set_ylabel('y', size=14, labelpad=-21, y=1.02, rotation=0)

ticks_frequency = 1
x_ticks = np.arange(xmin, xmax+1, ticks_frequency)
y_ticks = np.arange(ymin, ymax+1, ticks_frequency)
ax.set_xticks(x_ticks[x_ticks != 0])
ax.set_yticks(y_ticks[y_ticks != 0])

ax.set_xticks(np.arange(xmin, xmax+1), minor=True)
ax.set_yticks(np.arange(ymin, ymax+1), minor=True)

ax.grid(which='both', color='grey', linewidth=1, linestyle='-', alpha=0.2)

arrow_fmt = dict(markersize=4, color='black', clip_on=False)
ax.plot((1), (0), marker='>', transform=ax.get_yaxis_transform(), **arrow_fmt)
ax.plot((0), (1), marker='^', transform=ax.get_xaxis_transform(), **arrow_fmt)

# Definir puntos
punto1 = [-8, 7]
punto2 = [-2, 1]
punto3 = [-5,-2]
punto4 = [-1,5]

# podemos verificar el contenido de cada punto
# solo tecleando el nombre de la variable
# se puede notar que es un arreglo de dos valores
punto1

# también podemos verificar el tipo de variable
type(punto1)


# se agrupan las abcisas
x_values = [punto1[0], punto2[0]]
x1_values = [punto3[0], punto4[0]]

# también las ordenadas
y_values = [punto1[1], punto2[1]]
y1_values = [punto3[1], punto4[1]]

# tenemos un vector de abcisas
x_values

# también un vector de ordenadas
y_values

# podemos graficar ambos puntos
plt.scatter(x_values, y_values, color='green', marker= "*")
plt.scatter(x1_values, y1_values, color='purple', marker= "^")

# para graficar la línea entre los puntos
plt.plot(x_values, y_values, color = 'black')
plt.plot(x1_values, y1_values, color = 'orange')

# agregar etiquetas a puntos
# punto1[0] -> abcisa de punto 1
# punto1[1] -> ordenada de punto 1
plt.text(punto1[0]-0.015, punto1[1]+0.25, "A(-8,7)")
plt.text(punto2[0]-0.050, punto2[1]+0.35, "B(-2,1)")
plt.text(punto3[0]+0.110, punto3[1]-0.25, "C(-5,-2)")
plt.text(punto4[0]-0.450, punto4[1]+0.35, "D(-1,5)")

# Dibujar líneas de conexión entre los puntos y los ejes
colors = ['c', 'r', 'm', 'b']
for x, y, c in zip(x_values, y_values, colors):
    ax.plot([x, x], [0, y], c=c, ls='--', lw=1.5, alpha=0.5)
    ax.plot([0, x], [y, y], c=c, ls='--', lw=1.5, alpha=0.5)

colors = ['brown', 'violet', 'm', 'b']
for x, y, c in zip(x1_values, y1_values, colors):
    ax.plot([x, x], [0, y], c=c, ls='--', lw=1.5, alpha=0.5)
    ax.plot([0, x], [y, y], c=c, ls='--', lw=1.5, alpha=0.5)

plt.show()
```

```python
#Librerias
import numpy as np
import matplotlib.pyplot as plt
from pygeom import Axes, Point, Line

#definimos las lineas
linea1_abcisas = np.random.uniform(-20, 10, size=2)
linea1_ordenadas = np.random.uniform(-20, 10, size=2)
linea2_abcisas = np.random.uniform(-20, 10, size=2)
linea2_ordenadas = np.random.uniform(-20, 10, size=2)

# definimos algunas variables útiles
lim_ejes = (-20,10)
tam_fig = (11,11)

## Gráfica de una recta con dos puntos utilizando pygeom
# punto-1 de la línea 1 -> (x, y, color)
puntito1 = Point(linea1_abcisas[0], linea1_ordenadas[0], color='m')
puntito1.label = "Punto 1, Linea 1"
puntito2 = Point(linea1_abcisas[1], linea1_ordenadas[1], color='m')
puntito2.label = "Punto 2, Linea 1"

# punto-1 de la línea 2 -> (x, y, color)
puntito3 = Point(linea2_abcisas[0], linea2_ordenadas[0], color='deeppink')
puntito3.label = "Punto 1, Linea 2"
puntito4 = Point(linea2_abcisas[1], linea2_ordenadas[1], color='deeppink')
puntito4.label = "Punto 2, Linea 2"

# definimos la línea 1 entre los 2 puntos
linea1 = Line(p1=puntito1, p2=puntito2)
# definimos la línea 2 entre los 2 puntos
linea2 = Line(p1=puntito3, p2=puntito4)

# Crear los ejes cartesianos
ejes1 = Axes(xlim=lim_ejes, ylim=lim_ejes, figsize=tam_fig)

# agregar todos los elementos en el plano
ejes1.addMany([puntito1, puntito2, linea1, puntito3, puntito4, linea2])

# dibujar puntos y líneas
ejes1.draw()

# mostrar gráfico
plt.show()
```
