```python
#Librerias
from pygeom import Axes, Point, Line

#Instalacion paquete pyGeom2D usando pip
#!pip install pyGeom2D

## definimos los puntos
Punto1 = [-9, 5]
Punto2 = [-2, 8]

## definimos algunas variables útiles

# límite de ejes
lim_ejes = (-10,10)

# tamaño de figura
tam_fig = (13,13)

## Método 1: Utilizando dos puntos ##
## Gráfica de una recta con dos puntos utilizando pygeom

# punto-1 -> (x, y, color)
puntito1 = Point(Punto1[0], Punto1[1], color='m')
# punto-2
puntito2 = Point(Punto2[0], Punto2[1], color='deeppink')

# definimos la línea entre los 2 puntos
linea1 = Line(p1=puntito1, p2=puntito2)

# Crear los ejes cartesianos
ejes1 = Axes(xlim=lim_ejes, ylim=lim_ejes, figsize=tam_fig)

# agregar todos los elementos en el plano
ejes1.addMany([puntito1, puntito2, linea1])

# dibujar puntos y líneas
ejes1.draw()
   
```

```python
#Librerias
from pygeom import Axes, Point, Line

## Método 2: Utilizando un punto y la pendiente ##
## Gráfica de una recta con un punto y la pendiente utilizando pygeom

# Utilizamos el punto-1, pero ahora en azul
puntito1 = Point(Punto1[0], Punto1[1], color='blue')

# pendiente
pendiente = 1.0

# definimos la recta entre un punto y la pendiente
linea2 = Line(p=puntito1, slope=pendiente, color='green')

# Crear los ejes cartesianos
ejes2 = Axes(xlim=lim_ejes, ylim=lim_ejes, figsize=tam_fig)

# agregar todos los elementos en el plano
ejes2.addMany([puntito1, linea2])

# dibujar puntos y líneas
ejes2.draw()
    
```

```python
#Librerias
from pygeom import Axes, Point, Line

## Método 3: Utilizando la pendiente y la intersección en el eje ##
## Gráfica de una recta con un punto y la intersección en el eje y utilizando pygeom

linea3 = Line(slope=2.0, intercept=-1.0, color='gold')

# Crear los ejes cartesianos
ejes3 = Axes(xlim=lim_ejes, ylim=lim_ejes, figsize=tam_fig)

# agregar solo la línea en el plano
ejes3.add(linea3)

# dibujar puntos y líneas
ejes3.draw()
```

```python
#Librerias
import numpy as np
import matplotlib.pyplot as plt
from pygeom import Axes, Point, Line

#definimos las lineas
linea1_abcisas = np.random.uniform(-20, 10, size=2)
linea1_ordenadas = np.random.uniform(-20, 10, size=2)
linea2_abcisas = np.random.uniform(-20, 10, size=2)
linea2_ordenadas = np.random.uniform(-20, 10, size=2)

# definimos algunas variables útiles
lim_ejes = (-20,10)
tam_fig = (11,11)

## Gráfica de una recta con dos puntos utilizando pygeom
# punto-1 de la línea 1 -> (x, y, color)
puntito1 = Point(linea1_abcisas[0], linea1_ordenadas[0], color='m')
puntito1.label = "Punto 1, Linea 1"
puntito2 = Point(linea1_abcisas[1], linea1_ordenadas[1], color='m')
puntito2.label = "Punto 2, Linea 1"

# punto-1 de la línea 2 -> (x, y, color)
puntito3 = Point(linea2_abcisas[0], linea2_ordenadas[0], color='deeppink')
puntito3.label = "Punto 1, Linea 2"
puntito4 = Point(linea2_abcisas[1], linea2_ordenadas[1], color='deeppink')
puntito4.label = "Punto 2, Linea 2"

# definimos la línea 1 entre los 2 puntos
linea1 = Line(p1=puntito1, p2=puntito2)
# definimos la línea 2 entre los 2 puntos
linea2 = Line(p1=puntito3, p2=puntito4)

# Crear los ejes cartesianos
ejes1 = Axes(xlim=lim_ejes, ylim=lim_ejes, figsize=tam_fig)

# agregar todos los elementos en el plano
ejes1.addMany([puntito1, puntito2, linea1, puntito3, puntito4, linea2])

# dibujar puntos y líneas
ejes1.draw()

# mostrar gráfico
plt.show()
```
