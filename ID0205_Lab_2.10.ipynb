##[Polgono Convexo](https://es.wikipedia.org/wiki/Pol%C3%ADgono_convexo)

Un [polgono convexo](https://www.lifeder.com/poligono-convexo/) es un polgono en el que cada uno de los ngulos interiores miden a lo sumo 180 grados o  radianes. 

* Un polgono es estrictamente convexo si todos sus ngulos internos son estrictamente menores de 180 grados y todas sus diagonales son interiores. 

* Todo polgono que no es convexo se denomina polgono cncavo.

**Ejercicio 3**

* Usa los 10 puntos que has creado anteriormente.

* Muestra la lista usando 'print'
* Encuentra la poligonal que encierra esos puntos.
* Grafica la poligonal en color amarillo.
* Asigna un nombre a cada lnea (p. ej., Linea-0, Linea-1, etc.) que forma la poligonal y muestralos en la grafica.

```python
import numpy as np
import triangle as tr
import matplotlib.pyplot as plt

# Usa los 10 puntos que has creado anteriormente.
P = [(7, 4), (3, 2), (4, 2), (4, 5), (2, 2), (7, 2), (7, 6), (9, 5), (5, 8), (8, 4)]

# Mostrar la lista de puntos.
print("Lista de puntos:")
print(P)

# Encontrar la poligonal que encierra los puntos.
Vert = np.array(P)
poligonal = tr.convex_hull(Vert)

# Graficar la poligonal en color amarillo y asignar nombres a las lneas
fig, ax = plt.subplots()
for i in range(len(poligonal)):
    start_point = poligonal[i]
    end_point = poligonal[(i+1)%len(poligonal)]
    line_name = f"Linea-{i}"
    ax.plot(Vert[[start_point, end_point], 0], Vert[[start_point, end_point], 1], 
            color='yellow', label=line_name)

# Graficar los puntos originales
ax.plot(Vert[:, 0], Vert[:, 1], 'o', color='blue')

# Configuraciones de la grfica
ax.set_title("Poligonal convexa de los puntos")
ax.legend()
plt.show()
```

**Ejercicio 4**

* Lee el archivo 'dots' que viene en la librera de 'triangle'
* Encuentra la poligonal que encierra esos puntos.
* Grafica la poligonal en color cyan.
* Agrega un titulo a la figura.
* Agrega el cuadro en la parte superior derecha com la leyenda 'poligonal' donde aparezca la lnea y su leyenda.
* Cambia los marcadores de los puntos a tringulos verdes.

```python
import triangle as tr
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection

# Lee el archivo 'dots'
archivo_puntos = tr.get_data('dots')

# Obtn la lista de puntos
puntos = archivo_puntos['vertices']

# Encuentra la poligonal que encierra los puntos
segmentos = tr.convex_hull(puntos)

# Crea la instancia de LineCollection con los segmentos de la poligonal
lc = LineCollection([puntos[seg] for seg in segmentos], colors='cyan')

# Grafica la poligonal
plt.gca().add_collection(lc)

# Agrega un titulo a la figura
plt.title('Poligonal de los puntos')

# Agrega la leyenda 'poligonal' en la esquina superior derecha
plt.text(0.95, 0.95, 'Poligonal', ha='right', va='top', transform=plt.gca().transAxes)

# Cambia los marcadores de los puntos a tringulos verdes
plt.plot(puntos[:,0], puntos[:,1], 'gv')

# Muestra la figura
plt.show()
```

### Ejemplo 5

```python
## Ejemplo utilizando la librera scipy

# importa librera
from scipy.spatial import ConvexHull

# inicia grafica 
axis = plt.gca(); axis.cla();

# obtiene puntos de forma aleatoria
puntos_rand = np.random.rand(20, 2)

# grafica todos los puntos 
plt.plot(puntos_rand[:,0], puntos_rand[:,1], 'v')

# crea poligonal
poligonal = ConvexHull(puntos_rand)

# Get and plot the boundary
for simplex in poligonal.simplices:
    axis.plot(puntos_rand[simplex, 0], puntos_rand[simplex, 1], 'r-')

# muestra los vertices de la poligonal
print('Poligonal:\n ', np.array(poligonal.simplices), '\n\n')
```

**Ejercicio 5**

* Basandote en el ejemplo anterior define una funcin (metodo) en Python para que genere una poligonal usando de 5 a 20 puntos aleatorios.
* Crea un cdigo que llame la funcin que has creado.
* Usa una barra deslizante para permitir al usuario cambiar de 5 a 20 puntos aleatorios.
* Agrega un titulo a la figura.
* Grafica la poligonal en color brown.
* Agrega el cuadro en la parte superior derecha com la leyenda 'poligonal' donde aparezca la lnea y su leyenda.
* Cambia los marcadores de los puntos que NO son parte de la poligonal a cuadrados amarillos.
* Cambia los marcadores de los puntos que SI son parte de la poligonal a circulos verdes.

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, widgets

def generar_poligonal(num_puntos):
    x = np.random.rand(num_puntos)
    y = np.random.rand(num_puntos)
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.plot(x, y, color='brown', label='poligonal')
    ax.scatter(x[1:-1], y[1:-1], marker='s', color='yellow', label='puntos')
    ax.scatter(x[0], y[0], marker='o', color='green')
    ax.scatter(x[-1], y[-1], marker='o', color='green', label='inicio/fin')
    ax.set_title(f'Poligonal con {num_puntos} puntos')
    ax.legend(loc='upper right')
    plt.show()

interact(generar_poligonal, num_puntos=widgets.IntSlider(min=5, max=20, step=1, value=10));
```
