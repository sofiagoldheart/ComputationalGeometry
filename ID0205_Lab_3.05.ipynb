```python
# Import treeplot library
import treeplot as tree
import warnings
warnings.filterwarnings('ignore')
```

```python
# conjunto de datos disponibles en la librería
from sklearn import datasets
```

```python
# carga el conjunto de datos:'diabetes', 'digits', 'Linnerud', 'wine' and 'breast cancer'
data1 = datasets.load_diabetes()       #1
data2 = datasets.load_digits()         #2
data3 = datasets.load_linnerud()       #3
data4 = datasets.load_wine()           #4
data5 = datasets.load_breast_cancer()  #5
```

```python
# importa el algoritmo de árbol de decisiones
from sklearn.tree import DecisionTreeClassifier
```

```python
# aplica el árbol de decisiones para clasificar 'diabetes dataset'
model = DecisionTreeClassifier(max_depth=4, random_state=0).fit(data1.data, data1.target)
```

```python
# grafica el árbol resultante
ax = tree.plot(model, featnames=data1.feature_names)
```

```python
# aplica el árbol de decisiones para clasificar 'digits dataset'
model = DecisionTreeClassifier(max_depth=4, random_state=0).fit(data2.data, data2.target)
```

```python
# grafica el árbol resultante
ax = tree.plot(model, featnames=data2.feature_names)
```

```python
import graphviz
from sklearn.tree import DecisionTreeRegressor, export_graphviz
from sklearn import datasets
# aplica el árbol de decisiones para clasificar 'physical exercise Linnerud'
model = DecisionTreeRegressor(max_depth=4, random_state=0).fit(data3.data, data3.target)
```

```python
# grafica el árbol resultante
dot_data = export_graphviz(model, out_file=None, 
                           feature_names=data3.feature_names,  
                           filled=True, rounded=True,  
                           special_characters=True)
# Crear el grafo y visualizar el árbol de decisión
# Crear el grafo y mostrar el árbol de decisión
graph = graphviz.Source(dot_data) 
display(graph)
```

```python
# aplica el árbol de decisiones para clasificar 'physical exercise Linnerud dataset'
model = DecisionTreeClassifier(max_depth=4, random_state=0).fit(data4.data, data4.target)
```

```python
# grafica el árbol resultante
ax = tree.plot(model, featnames=data4.feature_names)
```

```python
# aplica el árbol de decisiones para clasificar 'physical exercise Linnerud dataset'
model = DecisionTreeClassifier(max_depth=4, random_state=0).fit(data5.data, data5.target)
```

```python
# grafica el árbol resultante
ax = tree.plot(model, featnames=data5.feature_names)
```

```python
# importa el algoritmo de Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier
```

```python
# aplica el Gradient Boosting para clasificar 'diabetes'
model = GradientBoostingClassifier().fit(data1.data, data1.target)
```

```python
# grafica el árbol resultante
# Exportar el primer árbol de decisión en formato Graphviz
dot_data = export_graphviz(model.estimators_[0][0], out_file=None, 
                           feature_names=data1.feature_names,  
                           filled=True, rounded=True,  
                           special_characters=True)

# Crear el grafo y mostrar el árbol de decisión
graph = graphviz.Source(dot_data) 
display(graph)
```

```python
# aplica el Gradient Boosting para clasificar 'physical exercise Linnerud'
model = GradientBoostingClassifier().fit(data3.data, data3.target[:, 0])
```

```python
# importa el algoritmo de xgboost
import xgboost as xgb
from sklearn import datasets
```
